```python
import tweepy
import json
from tweepy import OAuthHandler
import pandas as pd
```

```python
# Twitter provides REST APIs to interact with their service.
# Tweepy is Python-based client which is used to interact with Twitter and is easy to use.
# OAuth interface is used to authorise the app to access Twitter 
```

```python
auth=OAuthHandler(consumer_key,consumer_secret)
auth.set_access_token(access_token,access_secret)
api=tweepy.API(auth)
```

```python
#fetching all the tweets from the twitter handle @midasIIITD with 
#tweet_mode=extended helps in fetching the full text of the tweet.
#fetching these tweets and dump them in json format.
#putting each of these json formatted tweets to the JSONline file tweets.jsonl.
#JSONline file consists of each of these JSON formatted tweets in a newline.

#only fetches 140 characters of the tweet.----did not worked
# for s in tweepy.Cursor(api.user_timeline, id='midasIIITD').items():
#         print(s)
#not a very good method and also file is not in JSONline format.----did not worked
# for s in tweepy.Cursor(api.user_timeline, id='midasIIITD', tweet_mode="extended").items():
#     with open('tweet.json', 'a', encoding='utf8') as f:
#         json.dump(s._json, f)
```

```python
with open('tweets.jsonl', 'w', encoding='utf8') as f:
    for s in tweepy.Cursor(api.user_timeline, id='midasIIITD', tweet_mode="extended").items():
        json.dump(s._json, f)
        f.write('\n')
```

```python
#maintain a counter for the fetched tweets
i=0

#create new DataFrame to store all the tweets parsed from the JSONline file. 
tweets=pd.DataFrame(columns=['text','created_at','favourite_count','retweet_count','image_count'])

#open JSONline file and parse the tweets which are in JSON format line-by-line.
with open('C:/Users/Senjuti/Desktop/tweets.jsonl', 'r') as f:
    for JSONline in f:
        tweet = json.loads(JSONline)                  #load the JSON formatted twitter
        if 'extended_entities' not in tweet:          #used "entities" earlier but it fetches only the first media file
            countpic=None
        else:
            s=tweet['extended_entities']['media']
            countpic=0                               #maintain counter of the images.
            for j in s:
                if(j['type']=='photo'):              #if the 'type' of the media file in photo then increment the counter countpic
                    countpic=countpic+1
            
        #append the tweet to the tweets DataFrame.
        if ('retweeted_status' not in tweet) and ('RT @' not in tweet['full_text']):
            tweets.loc[i]=[tweet['full_text'],tweet['created_at'],tweet['favorite_count'],tweet['retweet_count'],countpic]
        else:
            #if the tweet is a retweet then fetch the full_text of the actual tweet and add the favorite count of the original tweet and the retweeted tweet
            tweets.loc[i]=[tweet['retweeted_status']['full_text'],tweet['created_at'],tweet['favorite_count']+tweet['retweeted_status']['favorite_count'],tweet['retweet_count'],countpic]
        #increment the counter of fetched tweets.
        i=i+1
```
